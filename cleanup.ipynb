```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\uwmap\Downloads\netflix_titles.csv")
print(df.head())
```

```python
#identifying missing values
print(df.isnull().sum())
```

```python
# adding unknown to missing values in 'director', 'cast', and 'country' columns
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
print(df.isnull().sum())
```

```python
# deleting rows where 'date_added' is null
df.dropna(subset=['date_added'], inplace=True)
print(df.isnull().sum())
```

```python
# Force to string, then strip, then convert to date
df['date_added'] = df['date_added'].astype(str).str.strip()
df['date_added'] = pd.to_datetime(df['date_added'])
```
